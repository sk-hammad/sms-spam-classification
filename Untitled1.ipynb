```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
df = pd.read_csv('spam.csv', encoding='latin1')
```

```python
df.head()
```

```python
df.drop(columns=['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], inplace=True) 
```

```python
df.head()
```

```python
df.rename(columns={'v1': 'class', 'v2':'text'}, inplace=True)
```

```python
df.head()
```

```python
df['text'] = df['text'].replace(to_replace=r'[^a-zA-Z]', value=' ', regex=True)
```

```python
df.head()
```

```python
word = []
alphabet = []
t = df['text']
for i in range(len(t)):
    alphabet.append(len(t[i]))
    word.append(len(t[i].split()))

print(word, alphabet)
```

```python
df["words"]=word
df["alphabet"]=alphabet
df
```

```python
sns.set(style="whitegrid")
class_counts = df["class"].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', colors=sns.color_palette("viridis"))
plt.title('class Distribution', fontsize=16)
plt.legend()
plt.show()
```

```python
spam_data = df[df['class'] == 'spam']
ham_data = df[df['class'] == 'ham']

numerical_columns = ['words', 'alphabet']

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))
for i, column in enumerate(numerical_columns):
    sns.histplot(spam_data[column], bins=20, kde=True, ax=axes[i], color='skyblue', label='Spam')
    sns.histplot(ham_data[column], bins=20, kde=True, ax=axes[i], color='salmon', label='Ham')
    axes[i].set_title(f'Histogram of {column}')
    axes[i].legend()
plt.tight_layout()
plt.show()

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))
for i, column in enumerate(numerical_columns):
    sns.boxplot(x='class', y=column, data=df, ax=axes[i], palette={'spam': 'skyblue', 'ham': 'salmon'})
    axes[i].set_title(f'Box Plot of {column}')
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
df['class'] = lb.fit_transform(df['class'])
```

```python
df.head()
```

```python
corpus = [] 
for i in range(0, len(df)):
  review = re.sub('[^a-zA-Z]', ' ', df['text'][i])
  review = review.lower()
  review = review.split()
  ps = PorterStemmer()
  all_stopwords = stopwords.words('english')
  all_stopwords.remove('not')
  review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
  review = ' '.join(review)
  corpus.append(review)
```

```python
df["new_text"] = corpus
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(corpus).toarray()
y = df['class'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB
snb = GaussianNB()
snb.fit(X_train, y_train)
```

```python
y_pred = snb.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score
print(confusion_matrix(y_pred, y_test))
print(accuracy_score(y_pred, y_test))
precision_score(y_pred, y_test)
```

```python
report_dic = {
    "Model": 'GaussianNB',
    "confusion matrix": confusion_matrix(y_pred, y_test),
    'accuracy score': accuracy_score(y_pred, y_test),
    'precision score': precision_score(y_pred, y_test)
}
report = pd.DataFrame([report_dic])
print(report)
```
